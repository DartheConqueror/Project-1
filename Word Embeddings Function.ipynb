```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize
import gensim
from gensim.models import word2vec
from gensim.models.keyedvectors import KeyedVectors
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
stem = pd.read_csv('df_stem.csv', encoding='latin-1')
not_stem = pd.read_csv('df_not_stemmed.csv', encoding='latin-1')
```

```python
embed_news = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin.gz", binary=True)
```

```python
def average_subjectivity(model, list_):
    subj_emb = [] #For all subjective words, getting words vectors from embed news model
    for i in list_:
        if i in model:
            subj_emb.append(model[i])#i is a subjective word in list, model[i] finds the vector for that word
        else: #creates a list of list
            continue
    avg_emb = np.asarray(sum(subj_emb)) #avg_emb = np.sum(subj_emb)#/len(list_) #/ length of list
    return avg_emb
```

```python
average_subjectivity(embed_news, stem)
```

```python
average_subjectivity(embed_news, not_stem)
```

```python
normalize([average_subjectivity(embed_news, not_stem)])
```
